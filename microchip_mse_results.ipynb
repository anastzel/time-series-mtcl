```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
typeA_mse_50um = [2.6903,
                  4.2320]
typeA_mse_20um = [3.1764,
                  4.6918]
typeB_mse_50um = [0.2019,
                  0.3063]
typeB_mse_20um = [0.2491,
                  0.3696]
typeC_mse_50um = [1.7758,
                  2.6133]
typeC_mse_20um = [3.1072,
                  3.9613]
typeD_mse_50um = [0.2136,
                  0.3111]
typeD_mse_20um = [0.2163,
                  0.2816]
typeE_mse_50um = [0.2634,
                  0.4267]
typeE_mse_20um = [0.2540,
                  0.3167]
```

```python
types = ["typeA", "typeB", "typeC", "typeD", "typeE"]
features = ["50μm", "20μm"]
archs = ["LSTM", "ARIMA"]

mses = {"typeA": [typeA_mse_50um, typeA_mse_20um],
        "typeB": [typeB_mse_50um, typeB_mse_20um],
        "typeC": [typeC_mse_50um, typeC_mse_20um],
        "typeD": [typeD_mse_50um, typeD_mse_20um],
        "typeE": [typeE_mse_50um, typeE_mse_20um]
        }
```

```python
fig, axs = plt.subplots(nrows=len(types), ncols=len(features), figsize=(9, 12))

for i in range(len(types)):
    feature_a, feature_b = features[0], features[1]
    
    # plot time signal:
    axs[i, 0].set_title(f"{types[i]} - {feature_a}")
    axs[i, 0].bar(archs,mses[types[i]][0], color=['blue', 'red'])
    axs[i, 0].grid(axis='y')

    # plot different spectrum types:
    axs[i, 1].set_title(f"{types[i]} - {feature_b}")
    axs[i, 1].bar(archs,mses[types[i]][1], color=['blue', 'red'])
    axs[i, 1].grid(axis='y')

fig.tight_layout()
plt.savefig("mses_results.png")
plt.show()
```
