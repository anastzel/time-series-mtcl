```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
# Copy the dataset.zip and unzip it
!cp "/content/drive/MyDrive/Microchip/dataset.zip" "/content"
!unzip dataset.zip
```

```python
def load_dataset(filename):
    # Read the .csv
    df = pd.read_csv(filename, delimiter=';')
    # Drop columns that we don't need
    df_cleaned = df.drop(df.columns[[0, 1, 2, 3]],axis = 1)
    # Set Dataframe's dtype to float32
    df_cleaned = df_cleaned.astype(np.float32)

    return df_cleaned
```

```python
types = ['type A', 'type B', 'type C', 'type D', 'type E']
```

```python
!mkdir /content/features_train_df/
!mkdir /content/features_test_df/
!mkdir /content/features_total_df/
```

## Plot Train Dataset

```python
for type in types:

    type_upd = type.replace(" ", "")

    train_filename = f"/content/dataset/{type_upd}/train/annotations.csv"
    train_df = load_dataset(train_filename)

    features = list(train_df.columns)

    feature_a, feature_b = features[0], features[1]

    fig, axs = plt.subplots(nrows=1, ncols=2, figsize=(15, 5))

    # plot time signal:
    axs[0].set_title(f"{type} - {feature_a}")
    axs[0].set_xlabel("Sample")
    axs[0].set_ylabel("Volume")
    axs[0].grid(True)
    axs[0].plot(train_df[feature_a], color='b', label='Train Data')
    axs[0].legend(loc ='upper left')

    # plot different spectrum types:
    axs[1].set_title(f"{type} - {feature_b}")
    axs[1].set_xlabel("Sample")
    axs[1].set_ylabel("Volume")
    axs[1].grid(True)
    axs[1].plot(train_df[feature_b],  color='b', label='Train Data')
    axs[1].legend(loc ='upper left')

    plt.savefig(f"/content/features_train_df/features_{type_upd}.png")

    
```

## Plot Test Dataset

```python
for type in types:

    type_upd = type.replace(" ", "")

    test_filename = f"/content/dataset/{type_upd}/test/annotations.csv"
    test_df = load_dataset(test_filename)

    features = list(test_df.columns)

    feature_a, feature_b = features[0], features[1]

    fig, axs = plt.subplots(nrows=1, ncols=2, figsize=(15, 5))

    # plot time signal:
    axs[0].set_title(f"{type} - {feature_a}")
    axs[0].set_xlabel("Sample")
    axs[0].set_ylabel("Volume")
    axs[0].grid(True)
    axs[0].plot(test_df[feature_a], color='b', label='Test Data')
    axs[0].legend(loc ='upper left')

    # plot different spectrum types:
    axs[1].set_title(f"{type} - {feature_b}")
    axs[1].set_xlabel("Sample")
    axs[1].set_ylabel("Volume")
    axs[1].grid(True)
    axs[1].plot(test_df[feature_b],  color='b', label='Test Data')
    axs[1].legend(loc ='upper left')

    plt.savefig(f"/content/features_test_df/features_{type_upd}.png")
```

## Plot Total Dataset

```python
for type in types:

    type_upd = type.replace(" ", "")

    train_filename = f"/content/dataset/{type_upd}/train/annotations.csv"
    test_filename = f"/content/dataset/{type_upd}/test/annotations.csv"
    train_df = load_dataset(train_filename)
    test_df = load_dataset(test_filename)
    test_df.index += len(train_df)

    features = list(train_df.columns)

    feature_a, feature_b = features[0], features[1]

    fig, axs = plt.subplots(nrows=1, ncols=2, figsize=(15, 5))

    # plot time signal:
    axs[0].set_title(f"{type} - {feature_a}")
    axs[0].set_xlabel("Sample")
    axs[0].set_ylabel("Volume")
    axs[0].grid(True)
    axs[0].plot(train_df[feature_a], color='b', label='Train Data')
    axs[0].plot(test_df[feature_a], color='r', label='Test Data')
    axs[0].legend(loc ='upper left')

    # plot different spectrum types:
    axs[1].set_title(f"{type} - {feature_b}")
    axs[1].set_xlabel("Sample")
    axs[1].set_ylabel("Volume")
    axs[1].grid(True)
    axs[1].plot(train_df[feature_a], color='b', label='Train Data')
    axs[1].plot(test_df[feature_a], color='r', label='Test Data')
    axs[1].legend(loc ='upper left')

    test_df.index -= len(train_df)

    plt.savefig(f"/content/features_total_df/features_{type_upd}.png")
```
